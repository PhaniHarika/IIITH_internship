# **TASK-4:Coffee Cup Detection Using YOLOv8n**

# **üéØ OBJECTIVE**

To detect and classify coffee cups and mugs using a custom-trained YOLOv8n model. This task includes:

1.Dataset preparation and annotation

2.Model training and validation

3.Inference on real-world images and frames

4.Visualization of prediction outputs

# **üß± PHASE 1: Environment Setup**

**1.1 Python & Virtual Environment**

python -m venv yolov8-env

cd yolov8-env/Scripts

activate

**1.2 Install Required Libraries**

pip install ultralytics opencv-python

# **üì¶ PHASE 2: Dataset Preparation**

**2.1 Annotated Dataset**

Downloaded a coffee-related video from YouTube using yt-dlp

Extracted frames using FFmpeg:

ffmpeg -i coffee_video.mp4 -vf "fps=1" frames/frame%04d.jpg

Annotated extracted frames using Roboflow with two labels: coffee mug and coffee cup

Exported dataset in YOLOv8 format

# **üß† PHASE 3: Model Training**

**3.1 Train the Model**

yolo task=detect mode=train model=yolov8n.pt data="path/to/data.yaml" epochs=30 imgsz=640 device=cpu

You can reduce epochs if needed

The data.yaml should list class names and paths to images/labels

**3.2 Resume Training (Optional)**

yolo task=detect mode=train resume=True

# **üìä PHASE 4: Evaluation**

**4.1 Validate Model Performance**
