# **Task-1:Image segmentation using YOLOv8**

**Explanation**

This task demonstrates how to apply instance segmentation on an image using the YOLOv8 model by Ultralytics. The objective is to not only detect objects in the image but also extract the exact shapes of the objects using segmentation masks.

**Prerequisites**:

VS Code installed → Download VS Code

Python installed → Download Python

Step 1: Create a Project Folder

Open VS Code.

Click File → Open Folder... and create a new folder called:

  YOLOv8_ObjectDetection_Internship
Inside that folder, create these files (right-click → New File):

  yolo_detect.py
Step 2: Set Up Python Virtual Environment

Open VS Code Terminal:

Click Terminal → New Terminal

In the terminal, type

 python -m venv venv
Activate it:

On Windows:

    venv\Scripts\activate
You’ll now see (venv) in the terminal — you're inside the virtual environment.

Step 3: Install YOLOv8 (Ultralytics)

Still in the terminal, install the Ultralytics YOLOv8 library:

pip install ultralytics
Step 4: Download the YOLOv8 Segmentation Model

Download the File Go to the official YOLOv8 model release:
Direct Link to Download

https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8s-seg.pt

  yolov8s-seg.pt
Right-click the link → Click "Save link as..."

Save the file to your computer (e.g., Downloads folder)

Move the File to Your Project Folder
Open your file explorer.

Locate the downloaded yolov8s-seg.pt file.

Drag and drop it into your project folder:.

Step 5: Write the Python Script

In yolo_detect.py, paste the following:

from ultralytics import YOLO

#Load the YOLOv8 segmentation model
model = YOLO('yolov8s-seg.pt')

# Run inference on an online sample image
results = model('https://ultralytics.com/images/bus.jpg')

# Show and save the output
results[0].show()   # Opens image viewer (desktop)
results[0].save()   # Saves to 'runs/segment/predict'
Step 6: Run the Script

In the terminal (with virtual environment active):

python yolo_detect.py
Step 7: View the Output

In VS Code:
Click on runs/segment/predict/ folder

Open bus.jpg

You’ll see:
Labeled bounding boxes
